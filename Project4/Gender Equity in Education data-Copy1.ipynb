# Mean Years in School (Men and women 25 and over)

The average number of years of school attended by all people in the age and gender group specified, including primary, secondary and tertiary education. 

Source: Institute for Health Metrics and Evaluation (IHME)

Link to complete reference:
http://www.healthmetricsandevaluation.org/resources/datasets/2010/education_attainment/education_attainment.html

```python
import pandas as pd
```

```python
mean_years_male = pd.read_csv('years_male.csv')
mean_years_fem = pd.read_csv('years_fem.csv')
```

```python
mean_years_male.sample()
```

```python
mean_years_male_clean = mean_years_male.copy()
mean_years_male_clean.info()
```

```python
mean_years_male_clean = pd.melt(mean_years_male_clean, id_vars= ['Country','UNESCO region'], var_name='year', value_name='mean years in school - Male')
```

```python
mean_years_fem.head()
```

```python
mean_years_fem_clean = mean_years_fem.copy()
mean_years_fem_clean = pd.melt(mean_years_fem_clean, id_vars= ['Country','UNESCO region'], var_name='year', value_name='mean years in school - female')
```

```python
mean_years_fem_clean.head()
```

```python
mean_years = pd.merge(mean_years_fem_clean,mean_years_male_clean, how='inner', 
                      on= ['Country', 'UNESCO region', 'year'])
```

```python
mean_years.head()
```

```python
mean_years= mean_years.drop('Row Labels', axis = 1)
```

```python
mean_years.tail()
```

```python
mean_years.to_csv('mean_years.csv')
```
